```python
plt.scatter(data_df["log_bm"], data_df["neo"])
plt.xlabel("Log-normalized Body Mass")
plt.ylabel("Neoplasia Prevalence")
```

```python
slope, intercept = np.polyfit(data_df["log_bm"], data_df["neo"], 1)
x_line = np.array([data_df["log_bm"].min(), data_df["log_bm"].max()])
y_line = slope * x_line + intercept

plt.scatter(data_df["log_bm"], data_df["neo"])
plt.ylim(0, 40)
plt.plot(x_line, y_line, color='red', linestyle='--')

plt.xlabel("Log-normalized Body Mass")
plt.ylabel("Neoplasia Prevalence")
```
